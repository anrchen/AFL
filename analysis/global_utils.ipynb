```python
import csv
import json

output_folder = '../../test_coverage20_data/test_coverage20_data_full/test_analysis/'

from itertools import islice
def take(n, iterable):
    "Return first n items of the iterable as a list"
    return list(islice(iterable, n))

from pathlib import Path
import os
def dict_to_json_wfolder(folder, file, dic):
    Path(os.path.join(output_folder, folder)).mkdir(parents=True, exist_ok=True)
    with open(os.path.join(output_folder, folder, file+'.json'), 'w') as fp:
        json.dump(dic, fp, sort_keys=True, indent=4)
    fp.close()
    
def dict_to_json(file, dic):
    with open(os.path.join(output_folder, file+'.json'), 'w') as fp:
        json.dump(dic, fp, sort_keys=True, indent=4)
    fp.close()
    
def dict_to_json_porder(file, dic):
    with open(os.path.join(output_folder, file+'.json'), 'w') as fp:
        json.dump(dic, fp, sort_keys=False, indent=4)
    fp.close()
    
def json_to_dict(file):
    data = {}
    with open(os.path.join(output_folder, file+'.json'), 'r') as fp:
        data = json.load(fp)
    fp.close()
    return data

def dict_to_csv(dict_data, csv_file, csv_columns, target_col):
    try:
        with open(csv_file, 'w') as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow(csv_columns)
            for k, data in dict_data.items():
                row = [k]
                for c in target_col:
                    row.append(data[c])
                writer.writerow(row)
    except IOError as e:
        print("I/O error", e)
        
def list_dict_to_csv(dict_data, csv_file, csv_columns, target_col):
    try:
        with open(csv_file, 'w', encoding='utf-8') as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow(csv_columns)
            for data in dict_data:
                row = []
                for c in target_col:
                    row.append(data[c])
                writer.writerow(row)
    except IOError as e:
        print("I/O error", e)

def create_csv(headers, file):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    ofile = os.path.join(output_folder, file+'.csv')
    if not os.path.exists(ofile):
        with open(ofile, 'a') as csvfile:
            c_writer = csv.writer(csvfile, delimiter=',', quotechar='"')
            c_writer.writerow(headers)
        
def writerow (rows, file):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    ofile = os.path.join(output_folder, file+'.csv')
    with open(ofile, 'a') as csvfile:
        c_writer = csv.writer(csvfile, delimiter=',', quotechar='"')
        for r in rows:
            c_writer.writerow(r)
            
def read_csv(file):
    ofile = os.path.join(output_folder, file+'.csv')
    with open(ofile, newline='', encoding='utf-8-sig') as csvfile:
        reader = csv.reader(csvfile, delimiter=',', quotechar='"')
        headers = next(reader, None)
        data = []
        for row in reader:
            i = 0
            item = {}
            for h in headers:
                item[h] = row[i]
                i += 1
            data.append(item)
            #print(item)
    return data

import re
def escape_ansi(line):
    ansi_escape = re.compile(r'(\x9B|\x1B\[)[0-?]*[ -\/]*[@-~]')
    return ansi_escape.sub('', line)

import subprocess
def execute_shell_file(file, args):
    # check_output waits for the output
    output = subprocess.check_output([file] + args, universal_newlines=True)
    return output

def execute_shell_file_poll(file, args):
    # invoke process
    process = subprocess.Popen([file] + args, stdin=subprocess.PIPE, stdout=subprocess.PIPE, universal_newlines=True)

    # Poll process.stdout to show stdout live
    output = ""
    counter = 0
    for stdout_line in iter(process.stdout.readline, ""):
        print('\r'+str(counter)+' lines processed ...', end='', flush=True)
        output += escape_ansi(stdout_line)
        counter += 1
    process.stdout.close()
    return_code = process.wait()
    if return_code:
        print('CalledProcessError | ' + file + ' | ' + ' '.join(args))
    return output
        
def execute_shell(args):
    process = subprocess.Popen(args,
                               stdout=subprocess.PIPE,
                               stderr=subprocess.PIPE,
                               universal_newlines=True)
    stdout, stderr = process.communicate()
    return stdout, stderr

def execute_shell_poll(args):
    process = subprocess.Popen(args,
                               stdout=subprocess.PIPE,
                               stderr=subprocess.PIPE,
                               universal_newlines=True)
    
    # Poll process.stdout to show stdout live
    output = ""
    counter = 0
    for stdout_line in iter(process.stdout.readline, ""):
        print('\r'+str(counter)+' lines processed ...', end='', flush=True)
        output += escape_ansi(stdout_line)
        counter += 1
    process.stdout.close()

    return_code = process.wait()
        
    return output

def execute_shell_cd(args, loc):
    process = subprocess.Popen(args,
                               stdout=subprocess.PIPE,
                               stderr=subprocess.PIPE,
                               universal_newlines=True,
                               cwd=loc)
    stdout, stderr = process.communicate()
    return stdout, stderr

def append(file, out):
    directory = output_folder+os.path.dirname(file)
    if not os.path.exists(directory):
        os.makedirs(directory)
    with open(output_folder+file+'.txt', 'a') as f:
        f.write(out)

def read_file_as_str(file):
    results = ""
    with open(file, 'r') as reader:
        # Read and print the entire file line by line
        for line in reader:
            results += line
    return results

def read_file_as_list(file):
    results = []
    ofile = os.path.join(output_folder, file+'.csv')
    with open(ofile, 'r') as reader:
        # Read and print the entire file line by line
        for line in reader:
            results.append(line.replace('\n', ''))
    return results

def write_to_file(folder, file, content):
    directory = folder+os.path.dirname(file)
    if not os.path.exists(directory):
        os.makedirs(directory)
    with open(os.path.join(folder, file), "w") as text_file:
        text_file.write(content)
        
def read_from_file(folder, file):
    directory = folder+os.path.dirname(file)
    if not os.path.exists(directory):
        print('dir', directory, 'does not exist')
        return
    return read_file_as_str(os.path.join(folder, file))
```
